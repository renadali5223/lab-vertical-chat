```python
#if you need an API Key from OpenAI
#https://platform.openai.com/account/api-keys

from openai import OpenAI
import os
from dotenv import load_dotenv, find_dotenv
_ = load_dotenv(find_dotenv())

OPENAI_API_KEY  = os.getenv('OPENAI_API_KEY')
```

```python
client = OpenAI(
    # This is the default and can be omitted
    api_key=OPENAI_API_KEY,
)

def continue_conversation(messages, temperature=0):
    response = client.chat.completions.create(
        model="gpt-3.5-turbo",
        messages=messages,
        temperature=temperature,
    )
    #print(str(response.choices[0].message["content"]))
    return response.choices[0].message.content
```

```python
def add_prompts_conversation(_):
    #Get the value introduced by the user
    prompt = client_prompt.value_input
    client_prompt.value = ''

    #Append to the context the User prompt.
    context.append({'role':'user', 'content':f"{prompt}"})

    #Get the response.
    response = continue_conversation(context)

    #Add the response to the context.
    context.append({'role':'assistant', 'content':f"{response}"})

    #Update the panels to show the conversation.
    panels.append(
        pn.Row('User:', pn.pane.Markdown(prompt, width=600)))
    panels.append(
        pn.Row('Assistant:', pn.pane.Markdown(response, width=600)))

    return pn.Column(*panels)
```

My Experience Creating a Chatbot with GPT-3.5 and Panel
I tried 3 different versions of the system prompt in a project called "Vertical Chat". The project is a chatbot that takes food orders in a virtual restaurant. Each version of the prompt gave the chatbot a different personality. I saw clear changes in the way the chatbot talked and answered.
First, I used a formal version. The chatbot was very professional. It always asked for details like drink sizes and toppings. It was good for a serious business or restaurant.
Second, I used a fun version. The chatbot was friendly and made funny comments. It was nice to talk to, but sometimes it forgot to ask important questions. This version is better for young people or social media.
Third, I used a healthy version. The chatbot gave advice about healthy food. It suggested salads or lighter drinks. It was helpful, but sometimes it repeated te advice too much, and it felt a bit annoying.
What I learned is that the prompt controls the chatbot personality. It decides how the chatbot talks and what it focuses on. Some versions made the chatbot give wrong or missing answers, like the fun one. So I learned that I need to find the right balance between personality and task.
